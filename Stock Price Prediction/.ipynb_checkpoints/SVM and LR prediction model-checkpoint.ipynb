```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import numpy as np
import pandas_datareader as web
```

```python
df = web.DataReader("RELIANCE.ns", "yahoo", start = "2014-01-01", end = "2020-03-24")
```

```python
df.head()
df = df[["Adj Close"]]
df
```

```python
#no. of days we want to predict
forecast_out = 1
#new column (target or dependent variable )
df["prediction"] = df[["Adj Close"]].shift(-forecast_out)
df
```

```python
#create the independent dataset x
#convert the df into numpy array
x = np.array(df.drop(["prediction"],1))
#remocve last n rows
x = x[:-forecast_out]
x
```

```python
#create the dependent dataset y
#convert the datafreame tp a numpy array 
y = np.array(df["prediction"])
#remocve last n rows
y = y[:-forecast_out]
y
```

```python
#split the data into train and test 
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
#create and train model (SVM)(regressor)
model = SVR(kernel = "rbf",C = 1e3, gamma = 0.1)
model.fit(x_train,y_train)

```

```python
#testing model 
# best possible score is 1.0
svm_confidence = model.score(x_test,y_test)
svm_confidence
```

```python
#create and train the linear regression model
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr_confidence = lr.score(x_test,y_test)
lr_confidence
```

```python
#set x_forecastequal to the last 30 rows of the original data set from adj, close column 
x_forecast = np.array(df.drop(["prediction"],1))[-forecast_out:]
x_forecast
```

```python
#print the LR model prediction for next"n" dasy
lr_prediction = lr.predict(x_forecast)
print(lr_prediction)
#print the LR model prediction for next"n" dasy
svm_prediction = model.predict(x_forecast)
print(svm_prediction)
```
